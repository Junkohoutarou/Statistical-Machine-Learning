<h4>1. MÔ HÌNH HỒI QUY TUYẾN TÍNH CƠ BẢN</h4>

```python
#Câu 1:
import numpy as np
import matplotlib.pyplot as plt

# Khởi tạo dữ liệu
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Thêm giá trị bias_term vào dữ liệu X ban đầu
X_b = np.c_[np.ones((100, 1)), X]

# Vẽ biểu đồ cho dữ liệu (X,y) vừa tạo ở trên.
plt.plot(X, y,"b.")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Biểu đồ dữ liệu (X, y)")
plt.show()
```

```python
#Câu 2: Cho biết chiều (shape) của X_b và y.
X_b.shape
```

```python
y.shape
```

```python
#Câu 3:Hãy viết code Python tìm ra tham số tối ưu cho mô hình hồi quy tuyến tính. 
XTX_inverse = np.linalg.inv(X_b.T.dot(X_b))#Sử dụng lệnh np.dot() trong thư viện numpy.
theta_best = XTX_inverse.dot(X_b.T).dot(y)
#Tạo ra dữ liệu mới X_new 
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
theta_best #Kết quả lưu vào biến theta_best.
```

```python
#Câu 4: Dùng tham số tối ưu của hồi quy tuyến tính vừa tìm được ở Câu 3, dự đoán kết quả cho X_new_b. 
y_pred = np.dot(X_new_b,theta_best)
y_pred #Giá trị dự đoán lưu vào biến y_pred.
```

```python
#Câu 5:Vẽ biểu đồ thể hiện kết quả dự đoán trên dữ liệu.
plt.plot(X, y, "b.", label="Dữ liệu huấn luyện")
plt.legend(loc="upper left")
plt.xlabel("X")
plt.ylabel("y")
plt.plot(X_new, y_pred, "r-",label="Linear Regression")
plt.legend(loc="upper left")
plt.title("Biểu đồ mô hình Linear Regression")
plt.show()
```

<h2>2. BÀI TẬP</h2>
<h3>Bài tập 4</h3>

```python
#Phần này tương tự bài 1
import numpy as np
import matplotlib.pyplot as plt

# Khởi tạo dữ liệu
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Thêm giá trị bias_term vào dữ liệu X ban đầu
X_b = np.c_[np.ones((100, 1)), X]

# Vẽ biểu đồ
plt.plot(X, y,"b.")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Biểu đồ dữ liệu (X, y)")
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X]
theta_best_l = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)#tính toán hệ số của mô hình hồi quy tuyến tính 
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_pred_l = X_new_b.dot(theta_best_l)
theta_best_l 
```

```python
#sử dụng thư viện scikit-learn để thực hiện hồi quy tuyến tính
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
y_sklearn_pred_l = lin_reg.predict(X_new)
```

```python
lin_reg.intercept_, lin_reg.coef_
```

```python
# Chuyển hệ số chặn intercept_ thành một ma trận hai chiều
intercept_matrix = np.array([lin_reg.intercept_])

theta_best_l_sklearn = np.concatenate((intercept_matrix, lin_reg.coef_), axis=0)
theta_best_l_sklearn
```

```python
theta_best_l
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(5, 4))
plt.plot(theta_best_l_sklearn, label="Scikit-Learn",  color="red", linewidth=2, marker="o", alpha=0.5)
plt.plot(theta_best_l, label="Manual",color="green", linewidth=1, marker="s", alpha=0.5)


plt.xlabel("Intercepts")
plt.ylabel("Coefficients")
plt.legend()
plt.title("So sánh tham số của Linear Regression bằng 2 phương thức")

# Sử dụng seaborn để làm đẹp biểu đồ
sns.set
plt.grid()
plt.show()
```

>Ta thấy cả hai cách tính hệ số đều cho kết quả tương tự,2 đường trùng nhau, nhưng thư viện scikit-learn giúp bạn thực hiện hồi quy tuyến tính một cách thuận tiện và hiệu quả hơn trong nhiều trường hợp.*--
