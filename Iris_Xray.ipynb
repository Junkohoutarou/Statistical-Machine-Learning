### Bài tập 2.
*Thực hiện huấn luyện mô hình KNN và SVM trên bộ dữ liệu, sau
đó so sánh độ chính xác (Accuracy) với mô hình LogisticRegression*

```python
#Import các thư viện cần thiết
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Đọc bộ dữ liệu từ thư viện Sklearn:
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data[:, :2] # Đối với X, chỉ sử dụng 2 thuộc tính sepal length và sepal width để dự đoán cho dữ liệu
y = iris.target # Nhãn, gồm 3 nhãn
```

```python
# Phân chia dữ liệu
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Sử dụng mô hình LogisticRegression trong thư viện sklearn để huấn luyện trên dữ liệu huấn luyện (X_train, y_train). Mô hình được lưu vào biến model.
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
# Dự đoán ra kết quả cho dữ liệu kiểm thử X_test và lưu vào biến y_pred. Sử dụng hàm predict():
y_pred = model_lr.predict(X_test)
```

```python
#Ta sử dụng độ chính xác (Accuracy) để đánh giá cho mô hình.
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# KNeighborsClassifier KNN
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=3, weights='distance')
knn_model.fit(X_train, y_train)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# SVM
from sklearn.svm import SVC

svm_model = SVC(kernel='linear',gamma='auto')
svm_model.fit(X_train, y_train)
```

```python
logistic_accuracy = model_lr.score(X_test, y_test)
knn_accuracy = knn_model.score(X_test, y_test)
svm_accuracy = svm_model.score(X_test, y_test)

print(f"Logistic Regression Accuracy: {round(logistic_accuracy * 100, 2)}%")
print(f"KNN Accuracy: {round(knn_accuracy * 100, 2)}%")
print(f"SVM Accuracy: {round(svm_accuracy * 100, 2)}%")
```

```python
models = ['Logistic Regression', 'KNN', 'SVM']
accuracies = [logistic_accuracy, knn_accuracy, svm_accuracy]

# Vẽ biểu đồ so sánh
plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['brown', 'yellow', 'violet'])
plt.title('So sánh độ chính xác của 3 mô hình')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Đặt giới hạn trục y từ 0 đến 1 cho độ chính xác

plt.show()
```

> Như vậy,sau khi thực hiện huấn luyện mô hình KNN và SVM trên bộ dữ liệu, sau đó đem so sánh độ chính xác (Accuracy) với mô hình LogisticRegression. Ta thấy, mô hình KNN có độ chính xác cao nhất (83.33%)

### Bài tập 3.
*Đánh giá 2 mô hình vừa xây dựng trên 3 độ đo sau:
precision_score, recall_score và f1_score sử dụng macro average.*

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Dự đoán của mô hình SVM và KNN trên tập dữ liệu kiểm tra
svm_pred = svm_model.predict(X_test)
knn_pred = knn_model.predict(X_test)

# Tính toán precision, recall và f1_score cho mô hình SVM và KNN bằng macro average
svm_precision = precision_score(y_test, svm_pred, average='macro')
svm_recall = recall_score(y_test, svm_pred, average='macro')
svm_f1 = f1_score(y_test, svm_pred, average='macro')

knn_precision = precision_score(y_test, knn_pred, average='macro')
knn_recall = recall_score(y_test, knn_pred, average='macro')
knn_f1 = f1_score(y_test, knn_pred, average='macro')

# In kết quả
print(f"SVM - Precision: {svm_precision * 100:.2f}%")
print(f"SVM - Recall: {svm_recall* 100:.2f}%")
print(f"SVM - F1 Score: {svm_f1* 100:.2f}%")
print("\n")
print(f"KNN - Precision: {knn_precision* 100:.2f}%")
print(f"KNN - Recall: {knn_recall* 100:.2f}%")
print(f"KNN - F1 Score: {knn_f1* 100:.2f}%")

```

```python
# tính toán bằng classification report
from sklearn.metrics import classification_report

# For SVM
svm_report = classification_report(y_test, svm_pred, target_names=['Setosa', 'Versicolour', ' Virginica'], zero_division=1)
print("SVM Classification Report:")
print(svm_report)

# For KNN
knn_report = classification_report(y_test, knn_pred, target_names=['Setosa', 'Versicolour', ' Virginica'], zero_division=1)
print("\nKNN Classification Report:")
print(knn_report)
```

### Bài tập 4.
*Sử dụng chiến lược tinh chỉnh tham số GridSearchCV để tìm ra
bộ tham số tốt nhất cho mô hình Logistic Regression. So sánh kết quả với
mô hình gốc.*

```python
# Import thư viện cần thiết
from sklearn.model_selection import GridSearchCV
import numpy as np

# Định nghĩa các giá trị của siêu tham số muốn thử nghiệm
param_grid = {
    "C": np.logspace(-10, 10, 100),  # Dãy giá trị logarit theo khoảng từ 10^-10 đến 10^10
    "solver": ["lbfgs"],  # Solver sử dụng cho Logistic Regression
    "penalty": ["l1", "l2"]
}

# Tạo một đối tượng GridSearchCV
lr_cv = GridSearchCV(model_lr, param_grid, cv=10)  # cv=10 là số lượng fold trong cross-validation

# Huấn luyện mô hình trên dữ liệu huấn luyện và tìm kiếm siêu tham số tốt nhất
lr_cv.fit(X_train, y_train)
```

```python
# In ra siêu tham số tốt nhất từ GridSearchCV
print("Siêu tham số được tinh chỉnh:", lr_cv.best_params_)

# In ra độ chính xác (accuracy) tốt nhất trên tập cross-validation
print("Accuracy:", round(lr_cv.best_score_ * 100, 2))
```

```python
# Đánh giá hiệu suất của mô hình Logistic Regression
logistic_accuracy = model_lr.score(X_test, y_test)
print(f"Accuracy của mô hình Logistic Regression: {round(logistic_accuracy * 100, 2)}%")

# Đánh giá hiệu suất của mô hình được tinh chỉnh
print(f"Accuracy của mô hình được tinh chỉnh: {round(lr_cv.best_score_ * 100, 2)}%")
```

>Kết quả này cho thấy rằng mô hình Logistic Regression sau khi được tinh chỉnh bằng Grid Search Cross Validation đã cải thiện độ chính xác so với mô hình gốc. Độ chính xác tăng lên từ 76.67% lên 82.5%.

<img src="https://i.imgur.com/jZqpV51.png" alt="Chest X-Ray ">


```python
import glob
import numpy as np
import cv2

IMG_SIZE = 227

def load_dataset(path):
    X = []
    y = []
    classes = ['NORMAL', 'PNEUMONIA']

    for c in classes:
        files = glob.glob(path + c + "/*.jpeg")
        for f in files:
            print(f)  # In ra tên file để theo dõi
            img = cv2.imread(f)

            # Kiểm tra xem ảnh có được đọc thành công không
            if img is not None:
                img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
                X.append(img)
                y.append(classes.index(c))  # Sử dụng chỉ mục của lớp thay vì chuỗi

    assert len(X) == len(y), 'Inconsistent number of samples between X and y'

    return (np.array(X), np.array(y))
```

```python

X_train, y_train = load_dataset("/content/drive/MyDrive/chest_xray/train/")
```

```python
X_test, y_test = load_dataset("/content/drive/MyDrive/chest_xray/test/")
```

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.imshow(X_train[0])
plt.title(" ảnh 1")

plt.subplot(1, 2, 2)
plt.imshow(X_test[0])
plt.title("ảnh 2")
plt.show()
```

### Đặc trưng HOG

```python
#câu a
from skimage.feature import hog
def feature_hog(img):
    fd, hog_image = hog(img, orientations=9, pixels_per_cell=(11, 11),\
                      cells_per_block=(2, 2), visualize=True, channel_axis=-1)
    return hog_image
```

```python
X3_hog = np.array([feature_hog(k) for k in X_test]).reshape(X_test.shape[0], IMG_SIZE*IMG_SIZE)
```

### Đặc trưng FAST

```python
#câu a
from skimage.feature import corner_fast
def feature_fast(img):
  fast_image = corner_fast(img, n=8, threshold=0)

  return fast_image
```

```python
X1_fast = np.array([feature_fast(k[:,:,0]) for k in X_train]).reshape(X_train.shape[0], IMG_SIZE*IMG_SIZE)
```
